```python
import numpy as np
import pyzx as zx
from fractions import Fraction
```

```python
c1=zx.Circuit(qubit_amount=5)
c1.add_gate("CNOT",0,1)
c1.add_gate("ZPhase", 2, phase=Fraction(3,4))
c1.add_gate("XPhase", 2, phase=Fraction(3,4))
c1.add_gates("HAD", 1)
c1.add_gate("T", 3)
c1.add_gate("S", 3)
c1.add_gate("T", 4,adjoint=True)
c1.add_gate("S", 4,adjoint=True)
zx.draw(c1)
```

```python
def add_CRZ(c,cq,tq,t):
    '''
    c: quantum circuit 
    cq: contol qubit
    tq: target qubit
    t: theta 
    '''
    
    c.add_gate("ZPhase", tq, phase=t/2)
    c.add_gate("CNOT",cq,tq)
    c.add_gate("ZPhase", tq, phase=-t/2)
    c.add_gate("CNOT",cq,tq)

c1=zx.Circuit(qubit_amount=5)
add_CRZ(c1,cq=2,tq=3,t=Fraction(3,4))
zx.draw(c1)
```

```python
def add_CRX(c,cq,tq,t):
    '''
    c: quantum circuit 
    cq: contol qubit
    tq: target qubit
    t: theta 
    '''
    c.add_gate("HAD", tq)
    c.add_gate("ZPhase", tq, phase=t/2)
    c.add_gate("CNOT",cq,tq)
    c.add_gate("ZPhase", tq, phase=-t/2)
    c.add_gate("CNOT",cq,tq)
    c.add_gate("HAD", tq)

c1=zx.Circuit(qubit_amount=5)
add_CRX(c1,cq=2,tq=3,t=Fraction(3,4))
zx.draw(c1)
```

```python
def add_CRY(c,cq,tq,t):
    '''
    c: quantum circuit 
    cq: contol qubit
    tq: target qubit
    t: theta 
    '''
    c.add_gate("S", tq ,adjoint=True)
    c.add_gate("HAD", tq)
    add_CRZ(c,cq,tq,t)
    c.add_gate("HAD", tq)
    c.add_gate("S",tq)

c1=zx.Circuit(qubit_amount=5)
add_CRY(c1,cq=2,tq=3,t=Fraction(3,4))
zx.draw(c1)
```

```python
def add_GZB(c,q1,q2,t):
    '''
    c: quantum circuit 
    cq: contol qubit
    tq: target qubit
    t: theta 
    '''
    '''
    add_GZB_matrix:
           [[1,0,0,0]
           [0,cos(t),sin(t),0]
           [0,sin(t),-cos(t),0]
           [0,0,0,-1]]
    '''

    c.add_gate("CNOT",q1,q2)
    c.add_gate("Z", q1 )
    add_CRY(c,q2,q1,np.pi-2*t)
    c.add_gate("CNOT",q1,q2)


c1=zx.Circuit(qubit_amount=5)
add_GZB(c1,q1=2,q2=3,t=Fraction(3,4))
zx.draw(c1)
```
