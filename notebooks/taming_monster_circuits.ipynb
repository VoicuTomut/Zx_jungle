```python
qubits = cirq.LineQubit.range(7)
circuit = cirq.Circuit()
#ri(rads=3.141592653589793).on(qubits[0])
circuit.append([X(qubits[1])])
circuit.append([S(qubits[2])])
circuit.append([Z(qubits[3])])
circuit.append([H(qubits[4])])
circuit.append(cirq.rz(np.pi*-0.15559005760408315).on(cirq.LineQubit(2)))
circuit.append(cirq.X(cirq.LineQubit(3)).controlled_by(cirq.LineQubit(2)))
circuit.append(cirq.Z(cirq.LineQubit(4)).controlled_by(cirq.LineQubit(3)))

print(circuit)
```
